```python
import pandas as pd
import yfinance as yf
```

```python
vix = pd.read_csv('VIX.csv')
vix = vix[['Date', 'Adj Close']]
vix.columns = ['Date', 'VIX_Close']
vix
```

```python
tnx = pd.read_csv('TNX.csv')
tnx = tnx[['Date', 'Adj Close']]
tnx.columns = ['Date', 'TNX_Close']
tnx
```

```python
tnx
```

```python
dollar_ind = pd.read_csv('Dollar Index.csv')
dollar_ind = dollar_ind[['Date', 'Adj Close']]
dollar_ind.columns = ['Date', 'Dollar_Close']
dollar_ind
```

```python
CPI = pd.read_csv('CPI Data.csv')
CPI.columns = ['Date', 'CPI_Close']
CPI
```

```python
GDP = pd.read_csv('GDP.csv')
GDP.columns = ['Date', 'GDP_Close']
GDP
```

```python
GSCI = pd.read_csv('GSCI.csv')
GSCI = GSCI[['Date', 'Adj Close']]
GSCI.columns = ['Date', 'GSCI_Close']
GSCI
```

```python
df  = vix.merge(tnx, on='Date', how='outer')
df = df.merge(dollar_ind, on='Date', how='outer')
df = df.merge(CPI, on='Date', how='outer')
df = df.merge(GDP, on='Date', how='outer')
df = df.merge(GSCI, on='Date', how='outer')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.sort_values('Date', inplace=True)
df.ffill(inplace=True)
df.reset_index(inplace=True, drop=True)
df = df.iloc[1:]
df
```
